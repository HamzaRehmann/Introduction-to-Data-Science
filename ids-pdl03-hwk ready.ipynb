## Assignment: Type conversion \[5\]<a href="#Assignment:-Type-conversion-%5B5%5D" class="anchor-link">¶</a>

Consider a string variable containing a large integer represented in
standard scientific notation (period is in the second place, first
character is non-zero) such as the following.

In \[ \]:

    strnum = '9.8765432109876543210987e25'

Direct conversion to integer type does not work. To take advantage of
the non-approximate integer arithmetics in Python, the string variable
must be decomposed and the integer representation reconstructed from the
partial strings.

Complete the code cell below according to the instructions included as
comments. The end result should be an integer representation of the
string above. You are advised to check the string function `split()`. A
convenient approach to constructing a string with `n` zeroes is to write
`n*'0'`. To verify that the code works, check with smaller integers
first.

In \[2\]:

    strnum = '9.8765432109876543210987e25'

    ### Split the string at the exponentation character 'e', store in variable lstnum.
    lstnum= strnum.split('e')
    #printing to check the output
    print(lstnum)

    ### Convert the second list element (index 1) from string to integer, store in variable pwrnum.
    pwrnum= int(lstnum[1])
    #printing to check the output
    print(pwrnum)

    ### Split the first list element (index 0) at the period '.', store in variable ampnum.
    s= lstnum[0]
    ampnum=s.split('.')
    print(ampnum)

    ### Store the number before the period (characteristic) and the fractional part (mantissa)
    ### in the variable chrnum and mntnum, respectively.
    chrnum=ampnum[0]
    mntnum=ampnum[1]

    ### Determine the length of the mantissa mntnum, store in variable lenmnt.
    lenmnt= len(mntnum)

    ### Construct a string with the correct number (= pwrnum-lenmnt) of zeros to be appended.
    appzrs= (pwrnum-lenmnt)*'0'

    ### Concatentate chrnum with mntnum and appzrs, then convert to integer.
    p=int(chrnum+mntnum+appzrs)
    print(p)

    25
    ['9', '8765432109876543210987']
    98765432109876543210987000

## Assignment: File handling \[6\]<a href="#Assignment:-File-handling-%5B6%5D" class="anchor-link">¶</a>

In the comma-separated text file `pdl03dat_SID.csv`, the first column is
time in the format hh:mm (hh: hour, mm: minute). The second column gives
temperature in Kelvin, with the numerical value `-99.99` indicating an
invalid measurement, i.e., a data gap.

Using the functions `open()` and `read()`, load the contents of the file
`pdl03dat_SID.csv`. Count the total number of data gaps. Change the time
format to hh.mm , i.e, replace the colon `:` with a period `.`.
Furthermore, substitute all occurences of the missing value `-99.99`
with the string `'NaN'`(not a number), and write the modified data to a
new file `pdl03new_SID.csv`. Note that the string function `replace()`
does not change the string variable itself but returns a copy.

In \[9\]:

    ### Test the code with SID=100, then use your personal student ID for the final run of the notebook.
    SID = 100
    ### Load file pdl03dat_SID.csv using open() and read() into string variable datcontent.
    pdl03dat = 'pdl03dat_'+str(SID)+'.csv'
    file=open(pdl03dat,'r')
    content=file.read()

    ### Close file pdl03dat_SID.csv.
    file.close()

    ### Count and print occurences of missing values.
    print("Occurences of missing values: {}".format(content.count('-99.99')))

    ### Replace '-99.99' with 'NaN', store in string variable newcontent.
    newcontent=content.replace('-99.99','NaN')

    ### In newcontent, change time format from hh:mm to hh.mm.
    newcontent=newcontent.replace('hh:mm','hh.mm')

    ### Open new file pdl03new_SID.csv for writing.
    pdl03new = 'pdl03new_'+str(SID)+'.csv'
    new_file=open(pdl03new,'r+')

    ### Write newcontent to file pdl03new_SID.csv.
    new_file.writelines(newcontent)

    ### Close file pdl03new_SID.csv.
    new_file.close()
